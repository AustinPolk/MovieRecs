```python
hyperparameters = {
    "UseAdjectives": True,
    "MinKeywordOccurrences": 10,
    "MaxKeywordOccurrences": 500,
    "EncoderHiddenDimensions": 64,
    "EncoderLatentDimensions": 8,
    "EncoderHiddenActivation": 'relu',
    "EncoderLatentActivation": 'relu',
    "DecoderHiddenDimensions": 64,
    "DecoderHiddenActivation": 'relu',
    "DecoderFinalActivation": 'sigmoid',
    "EncoderEpochs" : 75,
    "ClassifierMaxDepth": 5,
    "ClassifierEstimators": 10
}
```

```python
import os

dataFiles = []
for root, dirs, files in os.walk("training"):
    for file in files:
        path = os.path.join(root, file)
        dataFiles.append(path)
```

```python
from movie_reviews import MovieReviews

mr = MovieReviews(hyperparameters)

mr.train_encoder(dataFiles)
mr.set_user("Trevor")
mr.train_user_classifier()
```

```python
mr.EncodedMovieBank
```

```python
prediction, probability, regressed = mr.predict("Oppenheimer", -1)
print(prediction)
print(probability)
print(regressed)
```
